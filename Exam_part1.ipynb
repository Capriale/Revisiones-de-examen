##### 1. Search problem with graph
##### The Generator file contains code that generates a graph with 10 nodes starting from 0 to 10 and stores them in a list. The code generates the following elements:

* ##### Nodes: Numbers of the nodes start at zero and end at nine.
* ##### Edges: Relationship between the nodes “(#initial element, final element)”.
* ##### Degrees: Number of edges connected to that node

##### Generate a searcher in such a way that it provides the number of steps to get from 0 to 7.

```python
import networkx as nx
import matplotlib.pyplot as plt
import random


# Generation of a random graph
# Generacion de un grafo aleatorio
G = nx.gnm_random_graph(10, 15)


# Drawing the graph
# Dibujo del grafo
plt.figure(figsize=(8, 6))
nx.draw(G, with_labels=True, node_color='skyblue', node_size=700, edge_color='gray')
plt.title("Grafo Aleatorio")
plt.show()


# Acceso a la información del grafo
# Access to graph information
nodos = list(G.nodes())
aristas = list(G.edges())
grados = dict(G.degree())


print("Nodos del grafo:", nodos)
print("Aristas del grafo:", aristas)
print("Grado de los nodos:", grados)


# Ejemplo de vecinos de un nodo específico
# Example of neighbors of a specific node
nodo = 0  # Nodo específico
vecinos = list(G.neighbors(nodo))
print(f"Vecinos del nodo {nodo}:", vecinos)
```

```python
#Buscador BFS para encontrar el camino mas corto entre el nodo 0 y el nodo 7
def bfs_search(graph,inicio=0,fin=7):
    visitados=set()
    queue=deque([(inicio,0)]) 

    while queue:
        nodo_actual,pasos=queue.popleft()

        if nodo_actual==fin:
            return pasos

        if nodo_actual not in visitados:
            visitados.add(nodo_actual)

            for vecino in graph.neighbors(nodo_actual):
                if vecino not in visitados:
                    queue.append((vecino,pasos+1))
                    
    return None
```

```python
#Corremos el buscador para los nodos que escojamos
inicio=0
fin=7

pasos=bfs_search(G,inicio,fin))
print(f"Numero de pasos para llegar del nodo {inicio} al nodo {fin}: {pasos}")
```
