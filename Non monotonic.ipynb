```python
class LegalCase:
    def __init__(self, case_id, description, initial_evidence, initial_conclusion):
        self.case_id = case_id
        self.description = description
        self.initial_evidence = initial_evidence
        self.initial_conclusion = initial_conclusion
        self.new_evidence = None
        self.revised_conclusion = None

    #New evidence method
    def introduce_new_evidence(self, new_evidence):
        self.new_evidence = new_evidence

    #Conclusion revission using non-monotonic reasoning
    def revise_conclusion(self, revised_conclusion):
        self.revised_conclusion = revised_conclusion

    #Ccase details
    def display_case(self):
        print(f"Case ID: {self.case_id}")
        print(f"Description: {self.description}")
        print(f"Initial Evidence: {self.initial_evidence}")
        print(f"Initial Conclusion: {self.initial_conclusion}")
        if self.new_evidence:
            print(f"New Evidence: {self.new_evidence}")
            print(f"Revised Conclusion: {self.revised_conclusion}")
        print("")

cases = [
    LegalCase(
        case_id="Case 1: Theft in a Retail Store",
        description="A surveillance video shows an individual taking an item off the shelf and leaving the store without paying.",
        initial_evidence="Surveillance video showing theft.",
        initial_conclusion="The individual committed theft."
    ),
    LegalCase(
        case_id="Case 2: Breach of Contract",
        description="A company sues a contractor for failing to complete a project on time.",
        initial_evidence="Project deadline missed as per the contract.",
        initial_conclusion="The contractor breached the contract."
    ),
    LegalCase(
        case_id="Case 3: Wrongful Termination",
        description="An employee claims wrongful termination.",
        initial_evidence="Employee claims they were fired without cause.",
        initial_conclusion="Wrongful termination by the employer."
    ),
    LegalCase(
        case_id="Case 4: Self-Defense in an Assault Case",
        description="A person is charged with assault.",
        initial_evidence="Witnesses suggest the person was the aggressor.",
        initial_conclusion="The person is guilty of assault."
    ),
    LegalCase(
        case_id="Case 5: Inheritance Dispute",
        description="A will leaves the entire estate to one family member.",
        initial_evidence="A will is presented that distributes the estate to one person.",
        initial_conclusion="The will is valid."
    )
]
```

```python
#Introduce new evidence
cases[0].introduce_new_evidence("A witness states the individual left in a rush but returned to pay.")
cases[0].revise_conclusion("The act was an unintentional mistake, not theft.")

cases[1].introduce_new_evidence("Contractor provides evidence of force majeure (natural disaster).")
cases[1].revise_conclusion("Contractor is not liable due to force majeure.")

cases[2].introduce_new_evidence("Employer provides documented evidence of policy violations.")
cases[2].revise_conclusion("Termination was lawful based on policy violations.")

cases[3].introduce_new_evidence("Video footage shows the person acted in self-defense.")
cases[3].revise_conclusion("The person acted in self-defense, reducing charges.")

cases[4].introduce_new_evidence("A more recent will is found, redistributing the estate.")
cases[4].revise_conclusion("The new will supersedes the old one, distributing the estate differently.")
```

```python
#Display all case information
for case in cases:
    case.display_case()
```
