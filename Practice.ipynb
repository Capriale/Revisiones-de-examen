```python
import heapq
import networkx as nx
import matplotlib.pyplot as plt

# Grafo con heurística (valores arbitrarios)
graph = {
    'A': [('B', 1), ('C', 1)],
    'B': [('D', 2)],
    'C': [('D', 1)],
    'D': [('E', 3)]
}

# Poner los pesos estimados con heurística
heuristic = {
    'A': 20,
    'B': 15,
    'C': 10,
    'D': 5,
    'E': 0
}

def greedy_best_first_search(graph, start, goal):
    # Inicializar la cola de prioridad con el nodo de inicio
    priority_queue = [(heuristic[start], start, [start])]  # (h(n), nodo, camino)
    visited = set()  # Conjunto de nodos visitados

    while priority_queue:
        # Extraer el nodo con menor heurística
        _, node, path = heapq.heappop(priority_queue)

        # Si el nodo actual es el objetivo, retornar el camino
        if node == goal:
            return path

        if node not in visited:
            visited.add(node)  # Marcar el nodo como visitado
            for neighbor, _ in graph.get(node, []):
                if neighbor not in visited:
                    # Añadir el vecino a la cola de prioridad con su heurística
                    heapq.heappush(priority_queue, (heuristic[neighbor], neighbor, path + [neighbor]))

    return None  # No se encontró camino

# Ejecutar la búsqueda
start, goal = 'A', 'E'
path = greedy_best_first_search(graph, start, goal)
print("Camino encontrado por Greedy Best-First Search:", path)
```

```python
import heapq
import networkx as nx
import matplotlib.pyplot as plt

def a_star_tree_search(graph, start, goal):
    priority_queue = [(heuristic[start], 0, start, [start])]  # (f(n), g(n), nodo, camino)
    while priority_queue:
        _, g, node, path = heapq.heappop(priority_queue)
        if node == goal:
            return path
        for neighbor, cost in graph.get(node, []):
            new_g = g + cost
            f = new_g + heuristic[neighbor]
            heapq.heappush(priority_queue, (f, new_g, neighbor, path + [neighbor]))
    return None

def a_star_graph_search(graph, start, goal):
    priority_queue = [(heuristic[start], 0, start, [start])]  # (f(n), g(n), nodo, camino)
    visited = {}
    while priority_queue:
        _, g, node, path = heapq.heappop(priority_queue)
        if node == goal:
            return path
        if node not in visited or g < visited[node]:
            visited[node] = g
            for neighbor, cost in graph.get(node, []):
                new_g = g + cost
                f = new_g + heuristic[neighbor]
                heapq.heappush(priority_queue, (f, new_g, neighbor, path + [neighbor]))
    return None

# Ejecutar las búsquedas
start, goal = 'A', 'E'
path_tree = a_star_tree_search(graph, start, goal)
path_graph = a_star_graph_search(graph, start, goal)
print("Camino encontrado por A* Tree Search:", path_tree)
```
