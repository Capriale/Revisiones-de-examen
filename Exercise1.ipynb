#Cargar DataFrames

```python
file_name = 'iris.csv'
Iris = pd.read_csv(file_path + file_name)

Iris=Iris.sort_index()
print(Iris.head)
```

```python
file_name1 = 'Iris1.csv'
Iris1 = pd.read_csv(file_path + file_name1)

Iris1=Iris1.sort_index()
print(Iris1.head)
```

```python
file_name2 = 'Iris2.csv'
Iris2 = pd.read_csv(file_path + file_name2, on_bad_lines='warn')

Iris2=Iris2.sort_index()
print(Iris2.head)
```

```python
file_name3 = 'Iris3.csv'
Iris3 = pd.read_csv(file_path + file_name3)

Iris3=Iris3.sort_index()
print(Iris3.head)
```
