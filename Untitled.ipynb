## Parte 3

Grafo: \
A B 10 \
A C 15 \
B D 12 \
C D 10 \
D E 5 

Heurísticas:\
A E 20 \
B E 15 \
C E 10 \
D E 5 \
E E 0 

Al no haber nodos con ciclos con peso 0 y por ser un grafo pequeño, el algoritmo Uniform Cost Search es ideal para el problema. Otra opción más eficiente y que puede aprovechar el hecho de que tenemos heurísticas sería el algoritmo A*.

```python
from uniform_cost_search import UCS
from a_star_graph_search import ASGS
```

```python
graph = {
    'A': [('B', 10), ('C', 15)],
    'B': [('D', 12)],
    'C': [('D', 10)],
    'D': [('E', 5)],
}

start = 'A'
goal = 'E'

def heuristic(node, goal):
    heuristic_values = {
        'A': 20,
        'B': 15,
        'C': 10,
        'D': 5,
        'E': 0,
    }
    return heuristic_values[node]

print(UCS(graph, start, goal))
print(ASGS(graph, start, goal, heuristic))
```

El costo del camino es: 27.

### Comentario

Para este caso es más recomendable usar Uniform Cost Search debido al tamaño del grafo y porque el algoritmo A* podría haber tomado otro camino al considerar la suma de el costo acumulado con la heurística. En casos donde el grafo es más grande convendría usar algoritmos de busqueda informados.
